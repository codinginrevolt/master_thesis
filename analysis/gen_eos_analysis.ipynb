```python
n_end = 25
samples = np.load("/home/sam/thesis/code/results/eos_samples/archive/25ns_constraining_eos.npy")

#samples = np.load(f"/home/sam/thesis/code/results/eos_samples/{n_end}nsConnected_eos.npy")

ns = samples[0] #* 1.3234e-6
phis = samples[1] #* 1.3234e-6
#cs2s = samples[2]
```

```python
lowest = min(len(ns), 500)
# quantiles
q25 = np.percentile(phis, 25, axis=0)  
q50 = np.percentile(phis, 50, axis=0)  
q75 = np.percentile(phis, 75, axis=0)

plt.figure(figsize=(10, 6))
for i in range(len(ns)):
    plt.plot(ns[i], phis[i], color='gray', alpha=0.2)

plt.plot(ns[0], q25, label="25th Percentile", color='blue', linestyle='--')
plt.plot(ns[0], q50, label="50th Percentile", color='green', linestyle='-')
plt.plot(ns[0], q75, label="75th Percentile", color='red', linestyle='--')

# Fill the area between the 25th and 75th percentiles
plt.fill_between(ns[0], q25, q75, color='blue', alpha=0.1, label="Interquartile Range")

#plt.xlabel(r"n [n_sat]")
#plt.ylabel(r'$\phi$')

plt.xlabel(r"edens [MeV/fm^-3]")
plt.ylabel(r'pressure [MeV/fm^-3]')

plt.title(f"Successful Matching at {n_end}ns.")
plt.legend()
plt.grid(True)

#plt.xlim(0,2500)
#plt.ylim(0, 1200)

plt.show()
```

# function

```python
def plot_samples(n_end):
    """
    input n_end: ending number density value
    """
    samples = np.load(f"/home/sam/thesis/code/results/eos_samples/{n_end}nsConnecting_eos.npy")
    edens = samples[0]
    prezzes = samples[1]
    cs2s = samples[2]

    
    # quantiles
    q25 = np.percentile(prezzes, 25, axis=0)  
    q50 = np.percentile(prezzes, 50, axis=0)  
    q75 = np.percentile(prezzes, 75, axis=0)

    plt.figure(figsize=(10, 6))
    for i in range(len(edens)):
        plt.plot(edens[i], prezzes[i], color='gray', alpha=0.1)

    plt.plot(edens[0], q25, label="25th Percentile", color='blue', linestyle='--')
    plt.plot(edens[0], q50, label="50th Percentile", color='green', linestyle='-')
    plt.plot(edens[0], q75, label="75th Percentile", color='red', linestyle='--')

    # Fill the area between the 25th and 75th percentiles
    plt.fill_between(edens[0], q25, q75, color='blue', alpha=0.2, label="Interquartile Range")

    plt.xlabel(r"edens [MeV/fm^-3]")
    plt.ylabel(r'pressure [MeV/fm^-3]')

    plt.title(f"Samples with Successful Matching at {n_end}ns.")
    plt.legend()
    plt.grid(True)

    plt.xlim(0,2000)
    plt.ylim(0, 1200)

    plt.show()

```
