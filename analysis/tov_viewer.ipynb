````python
from scipy.interpolate import interp1d

n_TOV = np.zeros(100)

for i in range(100):
    p_i = p_c[i]

    valid_idx = np.where(~np.isnan(p_i))[0]
    
    if len(valid_idx) == 0:
        n_TOV[i] = np.nan 
        continue
    
    last_valid_p = p_i[valid_idx[-1]]

    # Interpolation for EOS i
    eos_interp = interp1d(p[i], n_test, bounds_error=False, fill_value="extrapolate")
    
    # Get n_TOV
    n_TOV[i] = eos_interp(last_valid_p)
````
