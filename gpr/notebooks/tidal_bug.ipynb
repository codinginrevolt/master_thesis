```python
def k2(C,y):
    term1 = (8*C**5/5) * (1-2*C)**2 * (2 + 2*C * (y - 1) - y )
    term2 = 2*C*(6 - 3*y + 3*C * (5*y - 8) )
    term3 = 4 * C**3 * (13 - 11*y  + C*(3*y - 2) + 2*C**2 * (1+y) )
    term4 = 3 * (1-2*C)**2 * ( (2 - y + 2*C*(y-1)) * np.log(1-2*C) )
    
    k2 = term1 * 1/(term2+term3+term4)

    return k2

def lam(k2, C):
    return 2/3 * k2 * C**-5
```
